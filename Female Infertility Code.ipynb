# Infertility In Female 
<p>Infertility in Female refers to the inability of a woman to conceive a pregnancy or carry a pregnancy to full term after actively trying for a year or more of regular, unprotected intercourse. It is a medical condition that affects the reproductive system and can be caused by various factors, including hormonal imbalances, structural abnormalities in the reproductive organs, ovulation disorders, pelvic inflammatory disease, endometriosis, and certain chronic illnesses. Infertility can be temporary or permanent, and it can be a result of a combination of factors or remain unexplained despite medical evaluation. It is important to seek medical advice and evaluation from a healthcare professional or fertility specialist for a proper diagnosis and potential treatment options.

Infertility prediction in Female using machine learning techniques is an area of active research and development. While it holds promise, it's important to note that accurately predicting infertility is a complex task that involves multiple factors and individual variability.

Machine learning algorithms can be trained on large datasets that include various features related to fertility, such as age, hormonal levels, medical history, lifestyle factors, and reproductive health indicators. By analyzing these features, machine learning models can learn patterns and correlations that may help in predicting the likelihood of infertility.
</p>

# Let me explain the columns in the dataset:

Patient ID: This column represents a unique identifier for each patient. It is used to distinguish one patient from another.

Age: This column captures the age of each patient. It indicates the age of the patient at the time the data was collected.

Ovulation Disorders: This column indicates whether a patient has ovulation disorders. Ovulation disorders refer to conditions that affect the regular release of eggs from the ovaries.

Blocked Fallopian Tubes: This column indicates whether a patient has blocked fallopian tubes. Blocked fallopian tubes can prevent the egg from reaching the uterus and can be a cause of infertility.

Endometriosis: This column indicates whether a patient has endometriosis. Endometriosis is a condition where the tissue lining the uterus grows outside of it, which can lead to infertility.

Uterine Abnormalities: This column indicates whether a patient has uterine abnormalities. Uterine abnormalities can include structural issues with the uterus that may affect fertility.

Pelvic Inflammatory Disease: This column indicates whether a patient has pelvic inflammatory disease. Pelvic inflammatory disease is an infection of the female reproductive organs that can cause infertility.

Hormonal Imbalances: This column indicates whether a patient has hormonal imbalances. Hormonal imbalances can disrupt the reproductive system and affect fertility.

Premature Ovarian Insufficiency: This column indicates whether a patient has premature ovarian insufficiency. It refers to a condition where the ovaries stop functioning normally before the age of 40, leading to infertility.

Autoimmune Disorders: This column indicates whether a patient has autoimmune disorders. Autoimmune disorders can affect fertility by causing the immune system to attack reproductive organs or cells.

Previous Reproductive Surgeries: This column indicates whether a patient has undergone previous reproductive surgeries. Past surgeries on the reproductive organs may impact fertility.

Unexplained Infertility: This column indicates whether a patient has unexplained infertility. Unexplained infertility refers to cases where no specific cause for infertility can be identified.

Infertility Prediction: This column represents the target variable that indicates whether a patient is predicted to be infertile or not. It is the variable you would typically want to predict or analyze using the other features/columns in the dataset.

# Train Test split

```python
# Importing the train_test_split function from sklearn.model_selection
from sklearn.model_selection import train_test_split

# Creating the predictors DataFrame by dropping the "Infertility Prediction" column
predictors = dataset.drop("Infertility Prediction", axis=1)

# Creating the target Series containing the "Infertility Prediction" column
target = dataset["Infertility Prediction"]

# Splitting the data into training and testing sets
# X_train: training set predictors, X_test: testing set predictors
# Y_train: training set target, Y_test: testing set target
X_train, X_test, Y_train, Y_test = train_test_split(predictors, target, test_size=0.20, random_state=0)
```

```python
# Retrieving the dimensions of the training set predictors
X_train.shape
```

```python
# Retrieving the dimensions of the testing set predictors
X_test.shape
```

```python
# Retrieving the dimensions of the training set target variable
Y_train.shape
```

```python
# Retrieving the dimensions of the testing set target variable
Y_test.shape
```

# Model Fitting

```python
# Importing the accuracy_score function from sklearn.metrics
from sklearn.metrics import accuracy_score
```

### Logistic Regression

```python
# Importing the LogisticRegression class from sklearn.linear_model
from sklearn.linear_model import LogisticRegression

# Creating an instance of the LogisticRegression model
lr = LogisticRegression()

# Fitting the logistic regression model on the training data
lr.fit(X_train, Y_train)

# Predicting the target variable for the testing data using the trained model
Y_pred_lr = lr.predict(X_test)
```

```python
# Retrieving the dimensions of the predicted target variable
Y_pred_lr.shape
```

```python
# Calculating the accuracy score using accuracy_score and rounding it to two decimal places
score_lr = round(accuracy_score(Y_pred_lr, Y_test) * 100, 2)

# Printing the accuracy score achieved using logistic regression
print("The accuracy score achieved using Logistic Regression is: " + str(score_lr) + " %")
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Assuming you have trained your Logistic Regression model and made predictions

# Generate classification report
report = classification_report(Y_test, Y_pred_lr)

# Print the classification report
print(report)

# Generate confusion matrix
cm = confusion_matrix(Y_test, Y_pred_lr)

# Print the confusion matrix
print("Confusion Matrix:")
print(cm)
```

### Random Forest

```python
# Importing the RandomForestClassifier class from sklearn.ensemble
from sklearn.ensemble import RandomForestClassifier

# Initializing variables
max_accuracy = 0

# Looping over different random states to find the best one
for x in range(2000):
    # Creating an instance of the Random Forest classifier with the current random state
    rf = RandomForestClassifier(random_state=x)
    
    # Fitting the Random Forest classifier on the training data
    rf.fit(X_train, Y_train)
    
    # Predicting the target variable for the testing data using the trained model
    Y_pred_rf = rf.predict(X_test)
    
    # Calculating the current accuracy score and rounding it to two decimal places
    current_accuracy = round(accuracy_score(Y_pred_rf, Y_test) * 100, 2)
    
    # Checking if the current accuracy is higher than the previous maximum accuracy
    if current_accuracy > max_accuracy:
        max_accuracy = current_accuracy
        best_x = x

# Creating a new instance of the Random Forest classifier using the best random state
rf = RandomForestClassifier(random_state=best_x)

# Fitting the Random Forest classifier on the training data with the best random state
rf.fit(X_train, Y_train)

# Predicting the target variable for the testing data using the trained model with the best random state
Y_pred_rf = rf.predict(X_test)
```

```python
# Retrieving the dimensions of the predicted target variable
Y_pred_rf.shape
```

```python
# Calculating the accuracy score using accuracy_score and rounding it to two decimal places
score_rf = round(accuracy_score(Y_pred_rf, Y_test) * 100, 2)

# Printing the accuracy score achieved using Random Forest
print("The accuracy score achieved using Random Forest is: " + str(score_rf) + " %")
```

```python
# Assuming you have trained your Naive Bayes Random Forest model and made predictions

# Generate classification report
report = classification_report(Y_test, Y_pred_rf)

# Print the classification report
print(report)
```

# ROC  of All Applied Classifiers on  Dataset

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# Assuming you have your training data (X_train, Y_train) and testing data (X_test, Y_test) for all classifiers

# Create an instance of the Logistic Regression model
lr = LogisticRegression()
# Fit the Logistic Regression model on the training data
lr.fit(X_train, Y_train)
# Predict the probabilities for the positive class for the testing data
lr_probs = lr.predict_proba(X_test)[:, 1]
# Compute the false positive rate (fpr), true positive rate (tpr), and thresholds for the ROC curve
lr_fpr, lr_tpr, _ = roc_curve(Y_test, lr_probs)
# Compute the area under the ROC curve (AUC)
lr_auc = auc(lr_fpr, lr_tpr)

# Create an instance of the Naive Bayes model
nb = GaussianNB()
# Fit the Naive Bayes model on the training data
nb.fit(X_train, Y_train)
# Predict the probabilities for the positive class for the testing data
nb_probs = nb.predict_proba(X_test)[:, 1]
# Compute the false positive rate (fpr), true positive rate (tpr), and thresholds for the ROC curve
nb_fpr, nb_tpr, _ = roc_curve(Y_test, nb_probs)
# Compute the area under the ROC curve (AUC)
nb_auc = auc(nb_fpr, nb_tpr)

# Create an instance of the SVM model
svm_model = SVC(kernel='linear', probability=True)
# Fit the SVM model on the training data
svm_model.fit(X_train, Y_train)
# Predict the probabilities for the positive class for the testing data
svm_probs = svm_model.predict_proba(X_test)[:, 1]
# Compute the false positive rate (fpr), true positive rate (tpr), and thresholds for the ROC curve
svm_fpr, svm_tpr, _ = roc_curve(Y_test, svm_probs)
# Compute the area under the ROC curve (AUC)
svm_auc = auc(svm_fpr, svm_tpr)

# Create an instance of the Random Forest classifier
rf = RandomForestClassifier(random_state=42)
# Fit the Random Forest classifier on the training data
rf.fit(X_train, Y_train)
# Predict the probabilities for the positive class for the testing data
rf_probs = rf.predict_proba(X_test)[:, 1]
# Compute the false positive rate (fpr), true positive rate (tpr), and thresholds for the ROC curve
rf_fpr, rf_tpr, _ = roc_curve(Y_test, rf_probs)
# Compute the area under the ROC curve (AUC)
rf_auc = auc(rf_fpr, rf_tpr)

# Plot the ROC curves
plt.figure(figsize=(12, 8))
plt.plot(lr_fpr, lr_tpr, label='Logistic Regression (AUC = %0.2f)' % lr_auc)
plt.plot(nb_fpr, nb_tpr, label='Naive Bayes (AUC = %0.2f)' % nb_auc)
plt.plot(svm_fpr, svm_tpr, label='SVM (AUC = %0.2f)' % svm_auc)
plt.plot(rf_fpr, rf_tpr, label='Random Forest (AUC = %0.2f)' % rf_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
